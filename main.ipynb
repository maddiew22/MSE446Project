```python
import numpy as np
import pandas as pd

df = pd.read_csv("data/AABA_2006-01-01_to_2018-01-01.csv")

df['Date'] = pd.to_datetime(df['Date'])
df = df.sort_values('Date')
df = df.set_index('Date')

# ensure continuous trading days
df = df.asfreq('B')

# backfill missing values
df = df.fillna(method="bfill")

#show missing values
print(df.isnull().sum())
```

```python
# Log returns which measure the percentage change in price from one day to the next
df["log_return"] = np.log(df["Close"] / df["Close"].shift(1))

# This is rolling volatility, which measures the standard deviation of log returns over a specified window (5, 20, and 60 days in this case). It captures the variability of returns and is often used as a measure of risk.
df["vol_5"] = df["log_return"].rolling(5).std()
df["vol_20"] = df["log_return"].rolling(20).std()
df["vol_60"] = df["log_return"].rolling(60).std()

# The range volatility is calculated as the difference between the high and low prices divided by the closing price. It provides a measure of the price range relative to the closing price, which can be an indicator of market volatility.
df["range_vol"] = (df["High"] - df["Low"]) / df["Close"]
```

```python
# Volume z-score, which measures how many standard deviations the current volume is from the mean volume over a rolling window (20 days in this case). It helps identify unusually high or low trading volumes.
df["vol_mean"] = df["Volume"].rolling(20).mean()
df["vol_std"] = df["Volume"].rolling(20).std()

df["volume_z"] = (df["Volume"] - df["vol_mean"]) / df["vol_std"]

df.head()

```
